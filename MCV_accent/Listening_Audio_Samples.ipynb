```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import torchaudio
import librosa
import IPython.display as ipd
import numpy as np
import json
```

```python
accents = ['philippines']
```

```python
# base_path = "/home/mayank/MTP/begin_again/Error-Driven-ASR-Personalization/MCV_accent/jsons/"
base_path = "/home/mayank/MTP/begin_again/Error-Driven-ASR-Personalization/MCV_accent/jsons/"

audio_invalidated = []
for accent in accents:
    json_path = f"{base_path}/{accent}.json"
    with open(json_path) as file:
        while True:
            line = file.readline()
            if not line:
                break
#             print(line.strip())
            line = json.loads(line.strip())
            audio_invalidated.append((accent, line["audio_filepath"], line["text"]))
index = np.random.choice(len(audio_invalidated))
print(f"accent = {audio_invalidated[index][0]}")            
print(f"text = {audio_invalidated[index][2]}")
speech, sr = torchaudio.load(audio_invalidated[index][1])
print(f"sampling rate is = {sr}")
speech = speech[0].numpy().squeeze()
speech = librosa.resample(np.asarray(speech), sr, 16_000)
ipd.Audio(data = np.asarray(speech), autoplay = True, rate = 16_000)
# file.close()
            
```
