```python
import numpy
import matplotlib.pyplot as plt
import json
import os
```

```python
def plot_hist(file_path):
    
    if not os.path.isfile(file_path):
        print(f"{file_path} is not a valid path.")
        return
    
    with open(file_path, 'r') as read_file:
        json_samples = [json.loads(_.strip()) for _ in read_file.readlines()]
        
#     print(len(json_samples), json_samples[:5])
    print(f"number of samples is = {len(json_samples)}")
    speaker_dict = dict()
    for json_sample in json_samples:
        speaker_id = json_sample["client_id"]
        
        if speaker_id in speaker_dict.keys():
            speaker_dict[speaker_id]+=1
        else:
            speaker_dict[speaker_id] = 1
    
    sorted_speakers = list(speaker_dict.keys())
    sorted_speakers.sort(key=lambda user_id: speaker_dict[user_id], reverse=True)
    
    sorted_counts = [speaker_dict[speaker_id] for speaker_id in sorted_speakers]
    
    plt.plot(sorted_counts)
    temp_ls = file_path.split('/')
    temp_ls = [_ for _ in temp_ls if len(_)>0]
    accent = temp_ls[1]
    method = temp_ls[-3]
    plt.title(f"Accent = {accent}, method = {method}")
    plt.show()
    
```

```python
file_path = 'data/african/all/budget_2500/target_25/FL2MI/39/budget_250/diverse_speakers/run_1/train.json'
plot_hist(file_path)
file_path = 'data/african/all/budget_2500/target_25/FL2MI/39/budget_250/random/run_1/train.json'
plot_hist(file_path)
```
