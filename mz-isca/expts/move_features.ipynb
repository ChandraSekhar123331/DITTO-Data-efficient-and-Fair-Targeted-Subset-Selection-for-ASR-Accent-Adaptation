```python
def is_same(source, target):
    source_lines = [json.loads(line.strip()) for line in open(source)]
    target_lines = [json.loads(line.strip()) for line in open(target)]
    if(len(source_lines) != len(target_lines)): return False
    if(len(source_lines) == 0): return True
    for src, trg in zip(source_lines, target_lines):
        contained = [(src[key] == trg[key] or (math.isnan(src[key]) and math.isnan(trg[key]))) for key in src]
        if not all(contained):
            # print(contained, [key for key in src]) 
            # print(src["gender"], type(src["gender"]))
            # print(src, trg)
            return False

    return True
```

```python
def move(source, target, target_dir):
    os.makedirs(target_dir, exist_ok=True)
    os.system(f"cp -v {source} {target}")
```
